# 예제 1

# Band Energy

```python
import numpy as np
import matplotlib.pyplot as plt
```

# constants

```python
h_bar = 1.05 * 10**(-34)
m_star = 9.11 * 10**(-31)
k = np.linspace(-np.pi, np.pi, 1000)
```

# Energy bands for conduction and valence bands

```python
E_conduction = (h_bar**2 * k**2)/(2*m_star)
E_valence = -E_conduction
```

# Plotting

```python
plt.plot(k, E_conduction, label='Conduction Band')
plt.plot(k, E_valence, label='Valence Band')
plt.xlabel(r'Wave Vector($k$)')
plt.ylabel(r'Energy($E$)')
plt.title('Energy Bands in a Semiconductor')
plt.legend()
plt.show()
```

# Si & Ge Energy Bands

### (연습) Si과 Ge의 energy bands 그래프 그려보기

```python
m_star_si = 0.26 * 9.11 * 10**(-31) 
m_star_ge = 0.56 * 9.11 * 10**(-31)

E_conduction_si = (h_bar**2 * k**2) / (2 * m_star_si)
E_valence_si = - E_conduction_si

E_conduction_ge = (h_bar**2 * k**2) / (2 * m_star_ge)
E_valence_ge = - E_conduction_ge


plt.plot(k, E_conduction_si, label='Conduction Band (Si)')
plt.plot(k, E_valence_si, label='Valence Band (Si)')
plt.plot(k, E_conduction_ge, label='Conduction Band (Ge)', linestyle='--')
plt.plot(k, E_valence_ge, label='Valence Band (Ge)', linestyle='--')
plt.xlabel(r'Wave Vector ($k$)')
plt.ylabel(r'Energy ($E$)')
plt.title('Energy Bands for Si and Ge')
plt.legend()
plt.show()
```
