# Carrier Concentration

```python
import math
import numpy as np
```

# Constants

```python
N_c = 2.8 * 10**(19)
N_v = 1.04 * 10**(19)
E_g = 1.12
k = 8.617 * 10**(-5)
```

# Temperature input

```python
T = float(input("Enter the temperature is Kelvin:"))
```

# Intrinsic carrier concenttration calculation

```python
n_i = math.sqrt(N_c*N_v)*math.exp(-E_g/(2*k*T))
n_i_np = np.sqrt(N_c*N_v)*math.exp(-E_g/(2*k*T))
```

# Output

```python
print(f"The intrinsic carrier concentration at {T} kelvin is {n_i:.2e} cm^-3.")
print(f"The intrinsic carrier concentration at {T} kelvin is {n_i_np:.2e} cm^-3.")
```

# 과제 - Si, Ge의 carrier concentration

```python
import math
```

```python
k = 8.617 * 10**(-5)
```

```python
N_c_Si = 2.8 * 10**19  
N_v_Si = 1.04 * 10**19
E_g_Si = 1.12

N_c_Ge = 1.0 * 10**19
N_v_Ge = 0.56 * 10**19 
E_g_Ge = 0.66 
```

```python
T = float(input("Enter the temperature is Kelvin:"))
```

```python
n_i_Si = math.sqrt(N_c_Si * N_v_Si) * math.exp(-E_g_Si/(2*k*T))
n_i_Ge = math.sqrt(N_c_Ge * N_v_Ge) * math.exp(-E_g_Ge/(2*k*T))
```

```python
print(f"The intrinsic carrier concentration at {T} kelvin is {n_i_Si:.2e} cm^-3.")
print(f"The intrinsic carrier concentration at {T} kelvin is {n_i_Ge:.2e} cm^-3.")
```

# 과제 - 온도에 따른 Intrinsic carrier concenttration의 변화 (로그스케일)

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
N_c = 2.8 * 10**19 
N_v = 1.04 * 10**19  
E_g = 1.12
k = 8.617 * 10**-5

# Temperature range
T = np.linspace(100, 1000, 100)

# Intrinsic carrier concenttration calculation
n_i = np.sqrt(N_c * N_v) * np.exp(-E_g / (2*k*T))

# Plotting
plt.plot(T, n_i, label="Intrinsic Carrier Concentration ($n_i$)")
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Temperature (K)')
plt.ylabel('Intrinsic Carrier Concentration ($n_i$)(cm$^{-3}$)')
plt.title('Intrinsic Carrier Concentration - Temperature')
plt.legend()
plt.show()
```
