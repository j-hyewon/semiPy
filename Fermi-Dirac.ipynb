# Fermi-Dirac Distribution

# 필요 라이브러리 등록

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 상수값들

```python
kB = 8.62*10**(-5) #볼츠만 상수 eV/K
```

```python
## Fermi-Dirac 함수 정의
```

```python
def FD_dist(E, E_f, T):
    f = 1/(1+np.exp((E-E_f)/(kB*T)))
    return f
```

# 에너지 값 목록

```python
Es = np.linspace(0, 3, 1000)
```

# 그래프 그리기

```python
plt.plot(Es, FD_dist(Es, 1, 100), label='T=100')
plt.plot(Es, FD_dist(Es, 1, 300), label='T=300')
plt.plot(Es, FD_dist(Es, 1, 1000), label='T=1000')
plt.xlabel(r'E(ev)')
plt.ylabel('Probability')
plt.title('Fermi-Dirac Distribution')
plt.legend()
plt.show()
```

# 온도 변경 후 그래프 그려보기

```python
plt.plot(Es, FD_dist(Es, 1, 50), label='T=50')
plt.plot(Es, FD_dist(Es, 1, 400), label='T=100')
plt.plot(Es, FD_dist(Es, 1, 2000), label='T=2000')
plt.xlabel(r'E(ev)')
plt.ylabel('Probability')
plt.title('Fermi-Dirac Distribution')
plt.legend()
plt.show()
```
